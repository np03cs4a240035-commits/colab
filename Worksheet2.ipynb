```python
#Task2
import pandas as pd

med_path = "/content/drive/MyDrive/Dataset/medical_students_dataset (1).csv"
df_med = pd.read_csv(med_path)

df_med.head()
```

```python
missing_cols = df_med.columns[df_med.isnull().sum() > 0]
missing_cols
```

```python
df_filled = df_med.copy()

for col in missing_cols:

    if df_med[col].dtype in ["int64", "float64"]:   # numeric column
        df_filled[col] = df_med[col].fillna(df_med[col].median())
        print(f"{col}  filled with median")

    else:                                           # categorical column
        df_filled[col] = df_med[col].fillna(df_med[col].mode()[0])
        print(f"{col}  filled with mode")
```

```python
df_filled.duplicated().sum()
```

```python
df_cleaned = df_filled.drop_duplicates()
df_cleaned.shape
```

```python
#Task3
import pandas as pd

titanic_path = "/content/drive/MyDrive/Dataset/Titanic-Dataset.csv"
df_titanic = pd.read_csv(titanic_path)

df_titanic.head()

```

```python
cols = ['Name', 'Pclass', 'Sex', 'Age', 'Fare', 'Survived']
df_sub = df_titanic[cols]
df_sub.head()
df_first = df_sub[df_sub['Pclass'] == 1]
df_first.head()
fare_mean = df_first['Fare'].mean()
fare_median = df_first['Fare'].median()
fare_max = df_first['Fare'].max()
fare_min = df_first['Fare'].min()

fare_mean, fare_median, fare_max, fare_min


```

```python
df_first['Age'].isnull().sum()

df_age_clean = df_first.dropna(subset=['Age'])
df_age_clean.head()
```

```python
df_encoded = df_titanic.copy()

dummy = pd.get_dummies(df_encoded['Embarked'], prefix="Embarked")
df_encoded = pd.concat([df_encoded, dummy], axis=1)
df_encoded.drop(columns=['Embarked'], inplace=True)

df_encoded.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

survival_by_sex = df_titanic.groupby('Sex')['Survived'].mean()
survival_by_sex

sns.barplot(x=survival_by_sex.index, y=survival_by_sex.values)
plt.title("Survival Rate by Sex")
plt.ylabel("Mean Survival Rate")
plt.show()

```

```python
df_plot = df_titanic[['Sex', 'Embarked', 'Survived']].dropna()

plt.figure(figsize=(8,5))
sns.barplot(data=df_plot, x='Embarked', y='Survived', hue='Sex')
plt.title("Survival Rate by Sex and Port of Embarkation")
plt.show()
```

```python
df_opt = df_titanic[['Age', 'Pclass', 'Survived']].dropna()
df_opt['AgeGroup'] = pd.qcut(df_opt['Age'], 5)

survival_stats = df_opt.groupby(['Pclass', 'AgeGroup'])['Survived'].mean()
survival_stats

plt.figure(figsize=(10,6))
sns.heatmap(
    survival_stats.unstack(),
    annot=True,
    cmap="Blues"
)
plt.title("Survival Rate by Age Group and Passenger Class")
plt.show()


```
